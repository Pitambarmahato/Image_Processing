```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread('a.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
#Locate XML Haar Cascade file
cascade_dir = 'C:/Users/pitam/Anaconda3/Lib/site-packages/cv2/data/'

#face cascade
face_cascade = cv2.CascadeClassifier(cascade_dir + 'haarcascade_frontalface_alt.xml')

#eye_cascade
eye_cascade = cv2.CascadeClassifier(cascade_dir + 'haarcascade_eye.xml')

#find bounding box coordinates of faces
faces = face_cascade.detectMultiScale(gray, 1.3, 5)

#loop through each faces and draw a rectangle 
for (x, y, w, h) in faces:
    img = cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), 2) #BGR format so the line is blue and 2 is the pixel size of the rectangle
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    eyes = eye_cascade. detectMultiScale(roi_gray)
    for (ex, ey, ew, eh) in eyes:
        cv2.rectangle(roi_color, (ex, ey), (ex+ew, ey+eh), (0, 255, 0), 2)
    
plt.imshow(img, cmap = 'gray')
```

```python
import cv2
cap = cv2.VideoCapture('http://192.168.1.64:8080/shot.jpg') #run camera

while True:
    ret, frame = cap.read()
    if ret:
        
        gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray_frame, 1.3, 5)

        for(x, y, w, h) in faces:
            frame = cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
            cv2.imshow('Frame', frame)
            k = cv2.waitKey(0) & 0XFF

            if k == 27: #code of escape key
                break
cap.release()
cv2.destroyAllwindows()
```
