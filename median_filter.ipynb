# Canny Edge Detector
* It is the best edge detection filtering mechanism.

```python
canny = cv2.imread('chess.jpg')
low = 10
high = 250

canny_filter = cv2.Canny(canny, low, high)
plt.imshow(canny_filter)
plt.show()
```
