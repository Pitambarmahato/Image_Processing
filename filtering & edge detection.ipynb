### What is image filtering????
* image filtering is the method of applying the filter to the image to make our image more useful.
* filters should have odd number of rows and columns
* sliding the filter over the image is known as the convolution
* the step of sliding the filter is known as stride
* the disadvantage of padding is that it will increase computation power
* we can use the same instead of padding 
* After Convolution step, we can apply the maxpooling to the convolved image. 

### Filtering Techniques
* High pass vs Low pass filtering
* low pass filter is used for smoothing and removes the noise of the image
* high pass filtering is used for edge detection.
### How can we find the edge
* if the pixel of image increases or decreases at a certain point then at that point the edge of the image is detected.
* after we apply the filter then the filter makes the image where the pixels values are damps and we define that as the edges
### Read about edge detection
* Apply high pass filtering

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
stripes = cv2.imread('chess.jpg', 0)
plt.imshow(stripes)

#Sobel_X
kernel1 = np.array([[-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1]])
filtered = cv2.filter2D(stripes, -1, kernel1)
plt.imshow(filtered)
plt.show()
```

* The above figure shows the vertical line detection of an image.

```python
kernel2 = kernel1.T
filtered = cv2.filter2D(stripes, -1, kernel2)
plt.imshow(filtered)
plt.show()
```

* Above figure detects the horizonal edges.

```python
kernel = kernel1 + kernel2
filtered = cv2.filter2D(stripes, -1, kernel)
plt.imshow(filtered)
plt.show()
```
